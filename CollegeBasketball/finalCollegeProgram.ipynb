```python
# need to save in a hashmap a team, and their stats
# then for each pair of teams, save the difference of favorite - non, home vs away, and the output
# want one to have only offense, one to have defense, one to have both, will start with both
    
import pandas as pd

# read the CSV file into a DataFrame
df = pd.read_csv('teamStats.csv')

# create an empty hashmap
data = {}

# iterate over the rows of the DataFrame
for index, row in df.iterrows():
    # access the first column of the row
    key = row.iloc[0]
    
    # access the rest of the columns of the row
    values = row.iloc[1:]
    
    # add the key and values to the hashmap
    data[key] = values
```

```python
# this is the lookup function given the two teams, returns a 2D array with the first teams stats, then the second
# print the hashmap
def difference(team1, team2):
    team2 = team2.strip()
    
    if team2 == "Morehouse" or team2 =='Southern Methodist' or team2 == 'Georgia Southwestern':
        return []
    
    if team2 == 'Mississippi - Ole Miss':
        team2 = "Ole Miss"
        
    if team2 == "Saint Mary's-California":
        team2 = "St. Marys"
        
    if team2 == 'North Carolina State':
        team2 = "NC State"
        
    team1Stats = [i for i in data[team1][1:-1]]
    team2Stats = [i for i in data[team2][1:-1]]
    # round each element to two decimal places
    diff = [round(x1 - x2, 2) for (x1, x2) in zip(team1Stats, team2Stats)]

    return diff
    
    
diff = difference('Marquette', ' Providence')
sum1 = round(sum(diff),2)
diff.append(sum1)
print(diff)
```

```python
# now need to get all final arrays, 
# this includes the difference, the sum, and the home vs away
# the target being if the second is W or L
finalList = []
for i in list2d:
        
    diff = []
    diff = difference(i[0], i[1])
    sum1 = round(sum(diff),2)
    diff.append(sum1)
    diff.append(i[9])
    
    if i[5] == "W":
        diff.append(1)
        finalList.append(diff)
    elif i[5] == "L":
        diff.append(0)
        finalList.append(diff)
    else:
        next
```

```python

# project architecture

# overview : predict the the outcome of a spread of a college basketball game with greater than 55% accuracy
    # branches : predicting with +-7 
    
# high-level : create a neural network that takes in an input of an array, and outputs a binary prediction

# low-level : training arrays : differential for statistics between two teams, and the outcome of the game
                            # : include home vs away (binary), H2H trends, previous trends
    

# steps --> get statistics for every single team, save in a hash map
# --> get a schedule for each team, then for a game, get the differentials, home v away, previous trends, H2H trends, (player injuries can be in the future, weighted #)
# --> get outcome of if they covered or not, then if they covered the 7 or not
# differentials = favorite - non favorite
# outcomes = if the favorite covered, if they covered the 7


# notes
    # can test win or loss fo fun
    # can also now have a winners and whiners or scp
    # for buying 1 point, 2, 3, start with 7
    # * do with score, then also with rankings

import tensorflow as tf
import numpy as np

# input dataset
X = np.array(finalAttributes)

# output dataset
y = np.array(finalTargets)

# define the model
model = tf.keras.Sequential([
  tf.keras.layers.Dense(3, activation=tf.nn.sigmoid),
  tf.keras.layers.Dense(1, activation=tf.nn.sigmoid)
])

# compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# train the model
model.fit(X, y, epochs=10000, verbose=0)

# evaluate the model
print(model.predict(X))
```

```python
import nltk

def getKey(team, data):
    # define the dictionary
    dictionary = data

    # define the target string
    target = team

    # create an empty list to store the matches
    matches = []

    # iterate over the keys and values in the dictionary
    
    for key, value in dictionary.items():
        # compute the edit distance between the key and the target string
        distance = nltk.edit_distance(key, target)
        # add the key and distance to the matches list
        matches.append((key, distance))

    # sort the matches by ascending distance
    matches.sort(key=lambda x: x[1])

    # print the key for the first (i.e., closest) match
    if matches:
        return matches[0][0]
```
